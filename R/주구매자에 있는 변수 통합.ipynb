```python
xdata = pd.read_excel('./데이터/2019 식품소비행태조사_성인가구원(배포용)_choice.xlsx', sheet_name = 'choice')

#범주형 변수도 함께 더함'SQ2_range','SQ1','I1_1_1_2'
#연속형 'A7','A8','A9','A16','B1',C4,I1_1_1_2,'BA1_1','BA1_2'

xdata = pd.concat([xdata, choicedata[['H_ID','A11_1','SQ2_range','A7','A9','A8','A16','B1','C4','I1_1_1_2','BA1_1','BA1_2','BA5','BA6','BA7']]],axis=1)
xdata
```

```python
#추가하고 싶은 변수

"""
H_id. 가구id *키값
SQ3N. 가구원 수
SQ3_1N. 동거 가구원 수
SQ7.OO님은 댁에서 직접 음식을 만들어드시나요?
A2_1. 식료품 구입 장소 (1순위) **범주형
A4. 온라인을 통한 식품 구입 빈도
B2. 쌀(백미, 현미) 조달 방법 **범주형
B3_1_1. 쌀(백미, 현미) 구입 시 우선 확인 정보 (1순위) **범주형
B4_1. 가장 많이 구입하는 잡곡 (1순위) **범주형
B5_1. 백미 구입 포장 단위 (_2:현미, _3:잡곡)  
B7. 유형별 쌀 구입 주기_씻어나온 쌀(수입, 친환경, GAP인증, 즉석밥) **범주형
B9_1. 주 취식 밥 종류 (1순위) ***y변수로 사용할 수 있음.
F19. 간편식(HMR)구입 빈도
BA1. 맞벌이 여부
BA2. 월평균 가구소득
BA2_1. 월평균 식료품 지출 비용
BA5_1. 식품 구입/조달 시 차량 이용 여부
K2. 청소년 가족 구성원 수
Dong. 동/ 읍면 구분 (도시와 시골 구분)"""
#범주형변수는 제외하고 추가하였음

avar = ['H_id','SQ3N','SQ3_1N','SQ7','A4','B5_1','B5_2','B5_3','F19','BA1','BA2','BA2_1','BA5_1','K2','Dong','A2_1','B2','B3_1_1']
```

```python
adata = mainbuyer[avar]
adata = adata.rename({'H_id':'H_ID'}, axis = 'columns')
adata
```

```python
Xdata = pd.merge(xdata,adata, on = 'H_ID', how = 'left')
Xdata
```

```python
# 'A11_1 (좋아하는 밥의 종류(1순위))' 데이터 전처리
for i in range(len(Xdata)):
    if Xdata['A11_1'][i] in [2, 3, 4]: 
        Xdata['A11_1'][i] = '잡곡'
    else:
        Xdata['A11_1'][i] = '백미'
```

```python
Xdata['A11_1'].value_counts()
```

```python
#각 열별 na데이터를 확인 후 na값이 많은 열은 삭제함.
Xdata = Xdata.dropna(axis = 1)
Xdata = Xdata.drop('H_ID',axis = 1)
Xdata
```
