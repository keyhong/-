```python
import requests
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time
import numpy as np
import datetime
import random
```

# 롯데마트

```python
browser = webdriver.Chrome('./chromedriver.exe')
names =[]
```

```python
url = "https://www.lotteon.com/search/search/search.ecn?render=search&platform=pc&q=%EC%9E%A1%EA%B3%A1&mallId=4" 
browser.get(url)
```

```python
#판매순으로 변경해주기
#다음장으로 넘겨 세번 반복
html = browser.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
items = soup.select('body div#app main > div > section > div.srchResultArea > section.srchResultContentArea ul div a div.srchProductUnitTitle')
```

```python
for i in range(60):
    names.append(items[i].text.replace('\n ','').replace('\n',''))
```

```python
#크롤링 결과 리스트로 저장
lottenames180 = names.copy()
```

## 이마트

```python
browser = webdriver.Chrome('c:/chromedriver.exe')
names =[]

url = 'http://emart.ssg.com/search.ssg?target=all&query=%EC%9E%A1%EA%B3%A1&sort=sale'
browser.get(url)
```

```python
#다음장으로 넘겨 두번 반복
html = browser.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
items = soup.select("body div div#container > div > div > div.tmpl_itemlist a > em.tx_ko")
```

```python
for i in items:
    names.append(i.text)
```

```python
#크롤링 결과 저장
ssgnames160 = names.copy()
ssgnames160
```

```python
lottenames180
```

```python
#롯데마트 상위150개와 쓱 상위150개 텍스트로 합치기

text1 = " ".join(ssgnames160[0:150])
text2 = " ".join(lottenames180[0:150])

text = text1+text2
text = text.replace('[맛있는 잡곡]','')
text
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus']=False
plt.rcParams["figure.figsize"] = (20,20) #사이즈설정

#필요없는 단어
stopword = ['1kg','2kg','500g','두보식품','맛있는 잡곡','4kg','900g','3kg','5kg','8kg','200g','7kg','210g','450g','800g','10kg','400g','CJ직배송']

#워드클라우드 만들기
wc = WordCloud(width=1000, height=600, background_color="white",stopwords = stopword , random_state=1, font_path=r'c:\Windows\Fonts\malgun.ttf')

#보여주기
plt.imshow(wc.generate(text))
plt.axis("off")
plt.savefig('제품명워드클라우드')
```
