```python
import requests
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time
import numpy as np
import datetime
import random
```

```python
browser = webdriver.Chrome('./chromedriver.exe')
grainurls = []
names =[]
```

```python
url = "https://terms.naver.com/tlist/list.nhn?listId=1070441&ownList=Y"
browser.get(url)
```

```python
#자료 다 나올수있게 v버튼 눌러준후 실행
html = browser.page_source
soup = BeautifulSoup(html, 'html.parser')
items = soup.select('span[class="__CARD_LINK"]')
```

```python
#각 잡곡별 url 목록 만들기.

for i in items:

    grainurls.append(i['href'])
    names.append(i.text)
```

```python
colnames=['니아신','나트륨','단백질','당질','레티놀','베타카로틴','비타민A','비타민 B','비타민 B2','비타민 B6','비타민C','비타민E','식이섬유','아연','엽산','인','지질','철분','칼륨','칼슘','콜레스테롤','회분']

#영양성분 저장할 빈 데이터 프레임 만들기
nutrient = pd.DataFrame(columns = colnames, index = names)
nutrient
```

```python
#영양성분 크롤링해서 데이터 프레임에 저장
a=0
for i in grainurls:
    browser.get(i)
    html = browser.page_source
    soup = BeautifulSoup(html, 'html.parser')
    items = soup.select('div.box_tbl')
    tmp = items[0].text.split()
    nutrient[colnames[0]][a] = tmp[0][3:-2]
    nutrient[colnames[1]][a] = tmp[1][3:-2]
    nutrient[colnames[2]][a] = tmp[2][3:-1]
    nutrient[colnames[3]][a] = tmp[3][2:-1]
    nutrient[colnames[4]][a] = tmp[4][3:-1]
    nutrient[colnames[5]][a] = tmp[5][5:-1]  
    nutrient[colnames[6]][a] = tmp[7][1:-3]
    nutrient[colnames[7]][a] = tmp[9][2:-2]
    nutrient[colnames[8]][a] = tmp[11][2:-2]
    nutrient[colnames[9]][a] = tmp[13][2:-2]
    nutrient[colnames[10]][a] = tmp[15][1:-2]
    nutrient[colnames[11]][a] = tmp[17][1:-2]
    nutrient[colnames[12]][a] = tmp[18][4:-1]
    nutrient[colnames[13]][a] = tmp[19][2:-2]
    nutrient[colnames[14]][a] = tmp[20][2:-1]
    nutrient[colnames[15]][a] = tmp[21][1:-2]
    nutrient[colnames[16]][a] = tmp[22][2:-1]
    nutrient[colnames[17]][a] = tmp[23][2:-2]
    nutrient[colnames[18]][a] = tmp[24][2:-2]
    nutrient[colnames[19]][a] = tmp[25][2:-2]
    nutrient[colnames[20]][a] = tmp[26][5:-2]
    nutrient[colnames[21]][a] = tmp[27][2:-1]
    time.sleep(3)
    a+=1
```
