## 품목명 분류 전처리

```python
# 백미를 담을 리스트
tmp_rice: List[str] = []

# 잡곡을 담을 리스트
tmp_grain: List[str] = [] 
```

#### 미곡

```python
rice = data[data['양곡분류구분']=='미곡'].reset_index()
rice.reset_index(drop=True, inplace=True)

tmp_rc = list(set(rice['품목명']))

# 찹쌀 및 잡곡 품목 확인 
for num in range(len(tmp_rc)):
    name = tmp_rc[num]    
    
    if '현미' in name:
        tmp_grain.append(name)
    
    else :
        tmp_rice.append(name)
```

#### 선물세트

```python
gift = data[data['양곡분류구분']=='선물세트'].reset_index()

tmp_gift = list(set(gift['품목명']))

for num in range(len(tmp_gift)):
    
    name = tmp_gift[num]
    
    if ('잡곡' in name) or ('혼합' in name):

        tmp_grain.append(name)

    else:
        if '한토래' in name:
            tmp_rice.append(name)
            
        else:
            tmp_grain.append(name)
```

#### PB상품

```python
# pb상품 중 쌀 포함된 품목 찾기
pb = data[data['양곡분류구분']=='PB상품'].reset_index()

tmp_pb = list(set(pb['품목명']))

# 쌀은'갯반닥서해일미 20kg','농협쌀(진공)500g' / 찹쌀제외 / 나머지 잡곡으로 확인

tmp_list = ['갯반닥서해일미 20kg', '농협쌀(진공)500g', '한토래고시히카리2kg(PET)/국산']

for num in range(len(tmp_pb)):
    
    name = tmp_pb[num]
    
    if ('현미' in name) or ('흑미' in name) or ('혼합' in name):
        tmp_grain.append(name)

    elif ('찹쌀' in name) or (name in tmp_list):
        tmp_rice.append(name)                            
    
    else:
        tmp_grain.append(name)
```
