```python
# Data Loading
data = pd.read_excel('../raw_data/농협양곡 매출데이터(2016년).xlsx')

# 정상만 인덱싱
data = data[data['정정취소구분']=='정상']
data.head()
```

## 품목명 분류 전처리

```python
# 백미를 담을 리스트
tmp_rice: List[str] = []

# 잡곡을 담을 리스트
tmp_grain: List[str] = [] 
```

#### 미곡

```python
rice = data[data['양곡분류구분']=='미곡'].reset_index(drop=True)
rice.reset_index(drop=True, inplace=True)

tmp_rc = list(set(rice['품목명']))

# 찹쌀 및 잡곡 품목 확인 
for num in range(len(tmp_rc)):
    name = tmp_rc[num]    
    
    if '현미' in name:
        tmp_grain.append(name)
        
    else:
        tmp_rice.append(name)
```

#### 선물세트

```python
gift = data[data['양곡분류구분']=='선물세트'].reset_index()

tmp_gift = list(set(gift['품목명']))

for num in range(len(tmp_gift)):
    
    name = tmp_gift[num]
    
    if name == '한토래발효잡곡밥6종세트3kg':
        tmp_grain.append(name)
    
    elif (name == '농협우리찹쌀식혜') or ('한토래' in name):
        tmp_rice.append(name)
        
    else:
        tmp_grain.append(name)
```

#### PB상품

```python
# pb상품 중 쌀 포함된 품목 찾기
pb = data[data['양곡분류구분']=='PB상품'].reset_index()

tmp_pb = list(set(pb['품목명']))

# 쌀은'갯반닥서해일미 20kg','농협쌀(진공)500g' / 찹쌀제외 / 나머지 잡곡으로 확인

for num in range(len(tmp_pb)):
    
    name = tmp_pb[num]
    
    if ('현미' in name) or ('흑미' in name) or ('잡곡' in name):
        tmp_grain.append(name)
        
    elif ('찹쌀' in name) and (name != '하나가득찹쌀+찰보리(각1kg)') or (name == '갯반닥서해일미 20kg') or (name == '농협쌀(진공)500g'):
        tmp_rice.append(name)
        
    else:
        tmp_grain.append(name)
```

#### 잡곡

```python
grain = data[data['양곡분류구분']=='잡곡'].reset_index()

tmp_gr = list(set(grain['품목명']))

for num in range(len(tmp_gr)):
    
    name = tmp_gr[num]

    if ('현미' in name) or ('흑미' in name) or ('잡곡' in name) or ('찹쌀보리' in name) or ('검정' in name):
        tmp_grain.append(name)
        
    elif ('햅쌀' in name) or ('올벼쌀' in name) or ('익산탑마루쌀' in name) or ('신동진' in name) or \
            ('당진해나루쌀' in name) or ('용의눈' in name) or ('선산쌀' in name) or ('찹쌀' in name) or ('예살비' in name):
        print(name)
        tmp_rice.append(name)
 
    elif ('싸라기' in name) or ('청미' in name):
        continue    
        
    elif '벼' in name:
        tmp_rice.append(name) 
    
    else:
        tmp_grain.append(name)
```

### 해당없음

```python
nn = data[data['양곡분류구분']=='해당없음'].reset_index()

tmp_nn = list(set(nn['품목명']))

tmp_list = [
    '15년산 일반벼', 'NH신동진 1kg', 'NH신동진 4kg', 'NH신동진 500g', 'NH신동진쌀 10kg',
    'NH신동진쌀 20kg', 'NH신동진쌀10kg', '승승장구 20kg', '신동진쌀 20kg', '예살비 10kg'
]

for num in range(len(tmp_nn)):
    
    name = tmp_nn[num]
    
    if name in tmp_list:
        print(name)
        tmp_rice.append(name)
 
    elif ('왕겨' in name) or ('청치(1kg)' == name):
        continue
        
    else:
        tmp_grain.append(name)
        print(name)
```

```python
tmp_rice = list(set(tmp_rice))
len(tmp_rice)
```

```python
tmp_grain = list(set(tmp_grain))
len(tmp_grain)
```

```python
data_copy = data.copy()
data_copy.loc[data_copy['품목명'].isin(tmp_rice), '품종'] = '미곡'
data_copy.loc[data_copy['품목명'].isin(tmp_grain), '품종'] = '잡곡'
data_copy
```
